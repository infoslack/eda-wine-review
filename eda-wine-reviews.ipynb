## Cleaning the data

```python
# make a copy
wines = df.copy()
```

```python
wines.columns
```

```python
# drop the missing columns
wines = wines.drop(["Unnamed: 0", "designation", "region_1", "region_2", "taster_name", "taster_twitter_handle"], axis=1)
wines.head()
```

```python
wines.info()
```

**Missing values in 'country', 'price', 'province', 'variety'**

```python
# proportion of price non null
print("{:.0%}".format(wines["price"].count() / wines.shape[0]))
```

**If we remove rows with null values in the price column this leaves 93% of the rows.
I'm comfortable removing these in such a big dataset.**

```python
# drop only the rows with missing values
wines = wines.dropna(subset=["country", "price", "variety"])
```

```python
wines.info()
```

## Exploratory data analysis

Questions we intend to answer with this analysis:

- Where does the best wine come from?
- Is price related to quality?
- Are certain varieties of wine higher quality?
- What is the best wine in each price bracket?
- What words are most used when describing wine?

```python
# check the distribution of points
plt.figure(figsize=(10, 7))
plt.hist(data=wines, x='points', bins=np.arange (80, wines['points'].max()+1, 1), color = 'purple')
plt.xlim(80,100)
plt.title('Dristribution of points', fontsize=14)
plt.xlabel('Points')
plt.ylabel('Frequency');
```

**There is approximately a normal distribution of points ranging from `80-100`**

```python
# top 10 wine producing
country_top_10 = wines["country"].value_counts()[:10]
country_top_10
```

```python
plt.figure(figsize=(10, 7))
country_top_10.plot(kind='bar', color = 'purple')
plt.title('Number of Wine Ratings by Country', fontsize = 14)
plt.xlabel('Country')
plt.ylabel('Number of Ratings');
```

**How many countries produce wine of the best quality according to our ratings ?**

```python
# top 10 countries with most average points
best_quality = wines.groupby("country").agg("mean")
best_quality = best_quality.sort_values(by="points", ascending=False)[:10]
best_quality.style.background_gradient(cmap='Purples',high=0.5, subset=["points"])
```

```python
y = best_quality.index.to_series()
x = best_quality["points"]

plt.figure(figsize=(10, 7))
plt.xlim(80, 92)
sns.barplot(x=x, y=y, palette='Purples_r')
plt.title('Top 10 Average Points by Country', fontsize = 14)
plt.xlabel('Average Points')
plt.ylabel('Country');
```

When looking at the highest average points per country it was `England` that produced the best quality wine. This is perhaps because there are fewer wines from `England` in the dataset and they have scored high, therefore, increasing the mean.

Let's investigating this further!

```python
wines["points"].describe()
```

```python
# let's look at those with scores over 95
high_scores = wines.drop(["description", "price", "province", "winery"], axis=1)
high_scores = high_scores[high_scores["points"] >= 95][:20]
high_scores.style.background_gradient(cmap='Purples',high=0.5, subset=["points"])
```

```python
# number of wines from each country scoring the highest ratings of 95-100 points
high_scores["country"].value_counts()
```

```python
# number of wines from each country
wines["country"].value_counts()
```

**The countries which achieved a rating of `95-100` points for their wines differ from our plot of the highest scores based on their means.**

A barplot will allow us to see the `max`, `min`, and `mean` scored by each country, which should paint a clearer picture of which countries score the highest points.

> *Since this dataset contains more US wines, it is likely that there will be a broad range of wines from the country, whereas England, which received the best average score, has fewer wines in the dataset, which is an unfair comparison.*

```python
plt.figure(figsize=(20, 18))
sns.boxplot(data=wines, x='points', y='country', color = 'Purple')
plt.title('Points by Country', fontsize = 14)
plt.xlabel('Points')
plt.ylabel('Country');
```

Although wines from `England` had the highest points on average, wines from Portugal, US, Italy, France, and Australia made wines that received a perfect score of `100`. There are many more wines from these countries in the dataset; therefore, they received a broader range of scores.

```python
# Now that we've seen the different distribution of quality of wine from each country,
# let's investigate the price of wine from other countries.

average_price = wines.groupby("country").agg("mean")
average_price = average_price.sort_values(by="price", ascending=False)
y = average_price.index.to_series()
x = average_price["price"]

plt.figure(figsize=(12, 8))
sns.barplot(x=x, y=y, palette='Purples_r')
plt.title('Average Price of Wine by Country', fontsize = 14)
plt.xlabel('Average Price ($)')
plt.ylabel('Country');
```

**Switzerland produces the most costly wine, with the average price being over `$80` for a bottle.
Wine from Ukraine is on the other end of the spectrum, with the average price being under just `$10`.**

> Is price related to quality of wine ?

```python
# first look at the distribution of wine prices
plt.figure(figsize=(10, 7))
bins = np.arange (0, wines["price"].max()+5, 5)
plt.hist(data=wines, x='price', bins=bins, color = 'Purple')
plt.xlim(0,200)
plt.title('Distribution of wine prices', fontsize=14)
plt.xlabel('Price($)')
plt.ylabel('Frequency');
```

**Price data is highly skewed to the right.
To better understand the distribution, we need a logarithmic transformation.**

```python
wines["price"].describe()
```

```python
np.log10(wines["price"].describe())
```

```python
# price distribution on a log scale
plt.figure(figsize=(12, 8))
bins = 10 ** np.arange(0.5, 3.5 + 0.05, 0.05)
ticks = [1, 3, 10, 30, 100, 300, 1000, 3000, 10000]
plt.hist(data=wines, x='price', bins=bins, color='Purple')
plt.title('Distribution of wine prices', fontsize=14)
plt.xscale('log')
plt.xticks(ticks, ticks)
plt.xlabel('Price ($)')
plt.ylabel('Frequency');
```

> There is a jump around the `$10` mark and a clear peak at just over `$20`.
There is another peak at wines priced at just over `$50`.

```python
# relationship between points and price
x = [1, 3, 10, 30, 100, 300, 1000, 3000, 10000]

plt.figure(figsize=(12, 8))
sns.regplot(x='price', y='points', data=wines, color='Purple', logx=True, scatter_kws={'alpha':0.3})
plt.title("Points vs. Price", fontsize=14)
plt.xscale('log')
plt.xticks(x, x)
plt.xlabel("Price($)")
plt.ylabel("Points");
```

**The scatterplot shows a positive relationship between points and price, with more expensive wines, tend to receive better ratings.**

```python
# the lowest priced wine
wines[wines["price"] == wines["price"].min()][:1]
```

```python
# the highest priced wine
wines[wines["price"] == wines["price"].max()][:1]
```

> What is the best wine in each price bracket?

- We now know that quality is related to price; how to get the best bang for your buck in each price range?

```python
# top 10 under $15
under_15 = wines.drop(["description", "province", "winery"], axis=1)
under_15 = under_15[under_15["price"] <= 15]
under_15 = under_15.sort_values(by="points", ascending=False)[:10]
under_15.style.background_gradient(cmap='YlOrRd',high=0.5, subset=["price"])
```

```python
# top 10 between $15 and $30
between_15_and_30 = wines.drop(["description", "province", "winery"], axis=1)
between_15_and_30 = between_15_and_30.query("price > 15 and price <= 30")
between_15_and_30 = between_15_and_30.sort_values(by="points", ascending=False)[:10]
between_15_and_30.style.background_gradient(cmap='YlOrRd',high=0.5, subset=["price"])
```

```python
# top 10 between $30 and $50
between_30_and_50 = wines.drop(["description", "province", "winery"], axis=1)
between_30_and_50 = between_30_and_50.query("price > 30 and price <= 50")
between_30_and_50 = between_30_and_50.sort_values(by="points", ascending=False)[:10]
between_30_and_50.style.background_gradient(cmap='YlOrRd',high=0.5, subset=["price"])
```

```python
# top 10 above $50
above_50 = wines.drop(["description", "province", "winery"], axis=1)
above_50 = above_50.query("price > 50")
above_50 = above_50.sort_values(by="points", ascending=False)[:10]
above_50.style.background_gradient(cmap='YlOrRd',high=0.5, subset=["price"])
```

**You can experience what 100 points tastes like for `$80`**

> WWhat words are most used to describe wine?

```python
# combine all wine reviews into one big text

from PIL import Image
from wordcloud import WordCloud, STOPWORDS

text = wines["description"][1]
text = " ".join(i for i in wines["description"])
len(text)
```

```python
# create stopword list
stopwords = set(STOPWORDS)
stopwords.update(["drink", "now", "wine", "flavor", "flavors"])
wordcloud = WordCloud(stopwords=stopwords, background_color="white", width=2048,height=1024, max_words=1000).generate(text)
```

```python
# display the generated image
plt.figure(figsize=(20,18))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off");
```
